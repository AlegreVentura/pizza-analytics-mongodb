```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_json('pizza.json')
df.tail()
```

```python
df['completed']=True
df
```

```python
# Efecto promocional: viernes de fin de mes fuertes + bajada a mediados de mes + San Valentin
df["order_date_parsed"] = pd.to_datetime(df["order_date"])
viernes_mask       = df["order_date_parsed"].dt.weekday == 4
dia_mes            = df["order_date_parsed"].dt.day
fin_mes_mask       = dia_mes >= 22   # ultimos 9 dias del mes
midmonth_mask      = dia_mes.between(10, 20)
diciembre_mask     = df["order_date_parsed"].dt.month == 12
valentin_mask      = (df["order_date_parsed"].dt.month == 2) & dia_mes.between(10, 16)

# +150% viernes de fin de mes (pico maximo semanal + mensual)
extra_vi_finmes    = df[viernes_mask & fin_mes_mask].sample(frac=1.5, replace=True, random_state=7)
# +60% viernes normales (no fin de mes)
extra_vi_normal    = df[viernes_mask & ~fin_mes_mask].sample(frac=0.6, replace=True, random_state=12)
# +80% viernes de diciembre fin de mes
extra_dic          = df[viernes_mask & diciembre_mask & fin_mes_mask].sample(frac=0.8, replace=True, random_state=8)
# +90% semana San Valentin
extra_valentin     = df[valentin_mask].sample(frac=0.9, replace=True, random_state=9)

df = pd.concat([df, extra_vi_finmes, extra_vi_normal, extra_dic, extra_valentin], ignore_index=True)

# Bajada a mediados de mes (dias 10-20): eliminar 40% de registros
midmonth_idx = df[df["order_date_parsed"].dt.day.between(10, 20)].index
drop_idx = df.loc[midmonth_idx].sample(frac=0.40, random_state=13).index
df = df.drop(index=drop_idx).reset_index(drop=True)

# Recalcular mascara despues del drop
df["order_date_parsed"] = pd.to_datetime(df["order_date"])

df = df.drop(columns=["order_date_parsed"])
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Reasignar IDs unicos para que drop_duplicates no elimine las filas extra
df = df.reset_index(drop=True)
df["order_details_id"] = range(1, len(df) + 1)
df["order_id"] = range(1, len(df) + 1)

print("Registros tras efecto promocional:", len(df))
```

```python
new_samples = df.sample(n=7100, replace=True, random_state=42)
new_samples["completed"] = False
```

```python
df = pd.concat([df, new_samples], ignore_index=True)  # Concatenar
df = df.sample(frac=1, random_state=42).reset_index(drop=True)  
```

```python
new_samples = df.sample(n=777, replace=True, random_state=42)

cols = new_samples.columns.drop('completed', errors='ignore')

# Probabilidad de que una celda sea NaN (ej: 10%)
prob_nan = 0.1

# Máscara aleatoria para seleccionar celdas
mask = np.random.random(size=new_samples[cols].shape) < prob_nan

# Asignar NaN a las celdas seleccionadas
new_samples.loc[:, cols] = new_samples[cols].mask(mask)

df = pd.concat([df, new_samples], ignore_index=True)  # Concatenar
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
```
